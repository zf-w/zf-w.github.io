```python
import altair as alt
import pandas as pd
import numpy as np
import geopy
```

```python
data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
df = pd.read_csv(data_url, na_values={"Square Footage": [0]})

df.columns
```

```python
df = df[["County", "Square Footage", "Year Acquired", "lat", "long"]]
df.head()
```

```python
df.dropna(inplace=True)
df_vis1 = df.groupby("County").agg('mean')['Square Footage']
vis1_index = list(df_vis1.index.values)
vis1_group = df.groupby("County")
vis1_footage_mean = [int(i) for i in vis1_group.agg('mean')['Square Footage'].values]
vis1_count = [int(i) for i in vis1_group.agg('count')['Square Footage'].values]


vis1_df = pd.DataFrame({"mean": vis1_footage_mean, "count": vis1_count})

vis2_df = pd.DataFrame({"x": df["County"], "year":df["Year Acquired"], "y": df["Square Footage"]})
vis2_df.head()
```

```python
brush = alt.selection_interval(encodings=['x'])

vis1_g1 = alt.Chart(vis1_df, title="Histogram of Mean of Square Footages of Counties", width=600).mark_bar().encode(
  alt.X("mean:Q", bin=alt.Bin(step=625), title="Mean Square Footage"),
  alt.Y('count()', title="Number of Counties")
).add_selection(
    brush
)
vis1_g2 = alt.Chart(vis1_df ,title="Histogram of Number of Buildings of Selected Counties", width=600).mark_bar().encode(
  alt.X("count:Q", bin=alt.Bin(step=5), title="Count"),
  alt.Y('count()', title="Number of Counties")
).transform_filter(
    brush
)

vis1 = alt.vconcat(vis1_g1, vis1_g2)
vis1
```

```python
vis2 = alt.Chart(vis2_df, width=600, title="Scatter Plot of Buildings' Square footage in Counties").mark_point().encode(
    alt.X("x:O"),
    alt.Y("y:Q", scale=alt.Scale(type="log")),
    alt.Color("year:Q", scale=alt.Scale(domain=[1800, 2100], range=["#000000", "#ffffff"]))
)

vis2
```
